```python
from PIL import Image
import os
import random
import csv

def generate_image_array(files, folder):
    dir_path = os.getcwd()
    directory = dir_path + folder
    for filename in os.listdir(directory):
        if filename == '.DS_Store':
            continue
        with Image.open(directory + filename) as im:
            files['filename'].append(filename)
            files['grape'].append([0, 0])
            files['leaf'].append([0, 0])
            files['vein'].append([0, 0])
            files['wire'].append([0, 0])

def check_images_size(folder):
    dir_path = os.getcwd()
    directory = dir_path + folder
    for filename in os.listdir(directory):
        if filename == '.DS_Store':
            continue
        with Image.open(directory + filename) as im:
            print(im.size)
```

```python
ROOT = "/data/redgrape/"
check_images_size(ROOT)
```

```python
ROOT = "/data/redgrape/"
files = dict(filename = [], grape = [], leaf = [], vein = [], wire = [])
generate_image_array(files, ROOT)
```

```python
with open('data.csv', 'wb') as f:
    writer = csv.writer(f)
    csv_header = ['filename', 'leaf', 'wire', 'grape', 'vein']
    writer.writerow(csv_header)
    for each in zip(files[csv_header[0]], files[csv_header[1]], files[csv_header[2]], files[csv_header[3]], files[csv_header[4]]):
        writer.writerow([each[0], each[1], each[2], each[3], each[4]])
```

```python
import pprint
 
def csv_dict_list(variables_file):
    reader = csv.DictReader(open(variables_file, 'rb'))
    data = []
    for line in reader:
        data.append(line)
    #for line in reader:
    #    for key, value in line.iteritems():
    #        data.setdefault(key,[]).append(value)
    return data

my_data = csv_dict_list('data_label.csv')
#pprint.pprint(my_data)
```

```python
from matplotlib.pyplot import imshow
import numpy as np
width =  200
height = 200
directory = os.getcwd() + "/data/redgrape/"

def centerCrop(im, x, y):
    left = np.ceil(x - width/2.)
    top = np.ceil(y - height/2.)
    right = np.floor(x + width/2.)
    bottom = np.floor(y + height/2.)
    sample = im.crop((left, top, right, bottom))
    return sample

raw_data = my_data[0:2]
for each_data in raw_data:
    for key, value in each_data.iteritems():
        if key == "filename":
            continue
        with Image.open(directory + each_data['filename']) as im:
            x = int(each_data[key].split(',')[0][1:])
            y = int(each_data[key].split(',')[1][:-1])
            sample = centerCrop(im, x, y)
            sample.save(os.getcwd()+"/data/training/"+key+"/"+each_data['filename'].split('.')[0]+"_"+key+".jpg")
```
